```python
import cv2 as cv
```

```python
mask_path = "D:\\Github\\JogosTeP\\Shadow\\Assets\\Material\\parede1\\parede1_MaskMap.png"
material_path = "D:\\Github\\JogosTeP\\Shadow\\Assets\\Material\\parede1"
material_name = "parede1"
```

```python
mask = cv.imread(mask_path, cv.IMREAD_UNCHANGED)
```

```python
mask_names = ["Metallic", "AmbientOcclusion", "DetailMask", "Smoothness"]
```

```python
import matplotlib.pyplot as plt
```

```python
for i in range(4):
    plt.title(mask_names[i])
    plt.imshow(mask[:,:,i])
    plt.show()
```

```python
def export(mask, material_name, path):
    for i in range(3):
        cv.imwrite(path+"//"+material_name+"_"+mask_names[i]+".png", mask[:,:,i])
    cv.imwrite(path+"//"+material_name+"_"+"Roughness"+".png", 1-mask[:,:,3])
```

```python
export(mask, material_name, material_path)
```
